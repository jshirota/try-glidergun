```python
from glidergun import grid
from IPython.display import clear_output


def tick(grid):
    g = grid.focal_sum() - grid
    return (grid == 1) & (g == 2) | (g == 3)


gosper = grid("input/gosper.txt")
# gosper = grid(
#     "input/LC08_L2SP_197021_20220324_20220330_02_T1_SR_B1.TIF").resample(4000).randomize() < 0.1

md5s = set()

while gosper.md5 not in md5s:
    md5s.add(gosper.md5)
    display(-(gosper := tick(gosper)))
    clear_output(wait=True)
```
